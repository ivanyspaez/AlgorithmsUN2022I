```python
import pydot
from collections import defaultdict 
from IPython.display import Image, display
import random
```

```python
def lstr(w):
        return ''.join([str(elem)  for elem in w])
```

```python
def Flip(s,q):
  list1 = s[0:q]
  list2 = list1[::-1]
  d = s
  if(q <= len(s)):
    for i in range(len(list2)):
      d[i] = list2[i]
    return d
```

```python
def Values(key, graph, visited, diagram):
  d = []
  graph[lstr(key)] = set()
  visited.add(lstr(key))
  node = pydot.Node(lstr(key))
  diagram.add_node(node)

  for n in range(2,len(key)+1):
    d = Flip(key.copy(), n)
    if not (lstr(d) in visited):
      visited.add(lstr(d))
      graph[lstr(key)].add(lstr(d))
      if d != sorted(d):
        node2 = pydot.Node(lstr(d))
      else:
        node2 = pydot.Node(lstr(d),style="filled", fillcolor="green")
      diagram.add_node(node2)
      edge = pydot.Edge(node,node2)
      diagram.add_edge(edge)
```

```python
def Convert(string):
    list1=[]
    list1[:0]=string
    return list1
```

```python
def Pancakes(s): 

  graph = {}
  visited = set()
  root = s
  answer = sorted(s)
  pn = 0
  G = pydot.Dot(graph_type="digraph")
  queue = []

  Values(s,graph,visited,G)
  queue.append(s)

  while queue:
    s = queue.pop(0)
    for i in graph[lstr(s)]:
      Values(Convert(i),graph,visited,G)
      queue.append(i)
  
  while True:
    pn += 1
    for i in graph.keys():
      for j in graph[i]:
        if lstr(answer) == j:
          answer = i
    if lstr(answer) == lstr(root):
      break


  print("Pn =",pn)
  im = Image(G.create_png())
  display(im)
```

```python
def CountValues(key, graph, visited):
  d = []
  graph[lstr(key)] = set()
  visited.add(lstr(key))
  node = pydot.Node(lstr(key))

  for n in range(2,len(key)+1):
    d = Flip(key.copy(), n)
    if not (lstr(d) in visited):
      visited.add(lstr(d))
      graph[lstr(key)].add(lstr(d))
```

```python
def PancakesPn(s): 

  graph = {}
  visited = set()
  root = s
  answer = sorted(s)
  pn = 0
  queue = []
  exit = False

  CountValues(s,graph,visited)
  queue.append(s)

  while queue:
    s = queue.pop(0)
    for i in graph[lstr(s)]:
      CountValues(Convert(i),graph,visited)
      queue.append(i)
      if(lstr(i) == lstr(answer)):
        exit = True
        break
    if exit:
      break
  
  while True:
    pn += 1
    for i in graph.keys():
      for j in graph[i]:
        if lstr(answer) == j:
          answer = i
    if lstr(answer) == lstr(root):
      break


  return pn
```

```python
def FlipTwoSpatulas(s,a,b):
  list1 = s[0:a]
  list2 = s[a:b]
  list3 = list2[::-1]
  d = s
  if(a <= len(s) and b <= len(s)):
    for i in range(len(list1)):
      d[i] = list1[i]
    for i in range(a,b):
      d[i] = list3[i-a]
  return d
```

```python
def ValuesDosespatulas(key, graph, visited, diagram):
  d = []
  graph[lstr(key)] = set()
  visited.add(lstr(key))
  node = pydot.Node(lstr(key))
  diagram.add_node(node)

  for n in range (0,len(key)+1):
    for m in range (n+1, len(key)+1):
      d = FlipTwoSpatulas(key.copy(), n, m)
      if not (lstr(d) in visited):
        visited.add(lstr(d))
        graph[lstr(key)].add(lstr(d))
        if d != sorted(d):
          node2 = pydot.Node(lstr(d))
        else:
          node2 = pydot.Node(lstr(d),style="filled", fillcolor="green")
        diagram.add_node(node2)
        edge = pydot.Edge(node,node2)
        diagram.add_edge(edge)
```
